```python
import numpy as np
import matplotlib.pyplot as plt
```

# Hamiltonian

```python
def hamiltonian(X0,t=0):
    return ([X0[1]/1.0,-5*(X0[0]-15)])
    
from scipy import integrate
t = np.linspace(0, 15,  100)              # time
#X0[0] is x, X[1] is p
X0 = np.array([1, 0])                     # initials conditions: 10 rabbits and 5 foxes
X, infodict = integrate.odeint(hamiltonian, X0, t, full_output=True)
```

```python
plt.plot(X[:,0])
plt.show()
plt.plot(X[:,1])
```
